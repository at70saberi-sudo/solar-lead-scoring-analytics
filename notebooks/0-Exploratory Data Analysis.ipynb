```python
df["lead_source_filled"] = df["lead_source"].fillna("Unknown")
```

```python
sns.barplot(
    data=df,
    x="lead_source_filled",
    y="purchased",
    errorbar=None
)
plt.title("Purchase Rate by lead source")
plt.xlabel("lead_source")
plt.ylabel("Purchase Rate")
plt.show()
```

```python
sns.scatterplot( 
    data= df,
    x = 'estimated_daily_energy_need_kwh',
    y = 'quoted_price_m_toman', 
    hue = 'purchased')
plt.title("Quoted Price vs Estimated Daily Energy Need")
plt.xlabel("Estimated Daily Energy Need (kWh/day)")
plt.ylabel("Quoted Price (million toman)")
plt.show()
          
```
